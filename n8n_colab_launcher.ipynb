# Run n8n on Google Colab (No Credit Card Required)

This notebook launches an **n8n automation server** inside Google Colab and exposes it using a **Cloudflare tunnel**.

Features:
- No payment or credit card required
- Persistent workflows using Google Drive
- Reuses existing n8n account if already created

Usage:

1. Run the Drive mount cell
2. Run the launcher cell
3. Open the printed public URL
4. Login using your existing n8n account

Notes:

- Colab sessions are temporary
- Public URL changes every run
- Workflows persist in Google Drive

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
!set -e

!echo "========== n8n Colab Launcher =========="

# Persistent storage
!export N8N_USER_FOLDER="/content/drive/MyDrive/n8n"
!mkdir -p "$N8N_USER_FOLDER"

!echo "Using storage: $N8N_USER_FOLDER"

#######################################
# Install Node.js if missing
#######################################

! if ! command -v node &> /dev/null
!then
!echo "Installing Node.js..."
!curl -fsSL https://deb.nodesource.com/setup_20.x | sudo -E bash -
!sudo apt-get install -y nodejs
!fi

!echo "Node version:"
!node -v

#######################################
# Install n8n if missing
#######################################

!if  command -v n8n &> /dev/null
!then
!echo "Installing n8n..."
!npm install -g n8n
!fi

!echo "n8n version:"
!n8n --version

#######################################
# Install Cloudflare tunnel if missing
#######################################

!if  command -v cloudflared &> /dev/null
!then
!echo "Installing Cloudflare tunnel..."
!wget -q https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64
!chmod +x cloudflared-linux-amd64
!sudo mv cloudflared-linux-amd64 /usr/local/bin/cloudflared
!fi

!echo "cloudflared version:"
!cloudflared --version

#######################################
# Start n8n
#######################################

!echo "Starting n8n server..."

!nohup n8n start > n8n.log 2>&1 &

!sleep 20

!echo "n8n started."

#######################################
# Create public tunnel
#######################################

!echo "Creating public tunnel..."

!cloudflared tunnel --url http://localhost:5678
```
